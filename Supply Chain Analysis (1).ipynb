```python
#Import Libraries & Load Data

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = '/Users/macpro/Downloads/supply_chain_data.csv'
df = pd.read_csv(file_path)

# Set visualization style
sns.set_style("whitegrid")
```

```python
df.head
```

```python
#Helper Function: Distribution Plot

def plot_distribution(data, column, title, xlabel, ylabel):
    plt.figure(figsize=(8, 5))
    sns.histplot(data[column], bins=20, kde=True, color="blue")
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
#Top Selling Products

top_products = df.groupby("Product type")["Number of products sold"].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
top_products.plot(kind="bar", color="skyblue")
plt.title("Top Selling Product Types")
plt.ylabel("Total Products Sold")
plt.xlabel("Product Type")
plt.xticks(rotation=45)
plt.show()
```

```python
#Inventory Optimization: Stock vs. Sales

plt.figure(figsize=(10, 5))
sns.scatterplot(x="Stock levels", y="Number of products sold", hue="Product type", data=df)
plt.title("Stock Levels vs. Number of Products Sold")
plt.xlabel("Stock Levels")
plt.ylabel("Products Sold")
plt.show()
```

```python
#Supplier Performance: Lead Times

plt.figure(figsize=(10, 5))
sns.boxplot(x="Supplier name", y="Lead time", data=df, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Supplier Lead Times Comparison")
plt.xlabel("Supplier")
plt.ylabel("Lead Time (days)")
plt.show()
```

```python
#Supplier Performance: Defect Rates

plt.figure(figsize=(10, 5))
sns.boxplot(x="Supplier name", y="Defect rates", data=df, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Supplier Defect Rates Comparison")
plt.xlabel("Supplier")
plt.ylabel("Defect Rate (%)")
plt.show()
```

```python
#Logistics: Shipping Times

plt.figure(figsize=(10, 5))
sns.boxplot(x="Transportation modes", y="Shipping times", data=df, palette="muted")
plt.title("Shipping Times by Transportation Mode")
plt.xlabel("Transportation Mode")
plt.ylabel("Shipping Time (days)")
plt.show()
```

```python
#Logistics: Shipping Costs

plt.figure(figsize=(10, 5))
sns.boxplot(x="Transportation modes", y="Shipping costs", data=df, palette="muted")
plt.title("Shipping Costs by Transportation Mode")
plt.xlabel("Transportation Mode")
plt.ylabel("Shipping Cost (£)")
plt.show()
```
